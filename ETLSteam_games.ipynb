```python
# Función para verificar si el valor es numérico o no
def convert_to_zero_if_non_numeric(value):
    if value is None or isinstance(value, str) and not value.replace('.', '', 1).isdigit():
        return 0
    try:
        # Intentar convertir el valor a un float si no es None
        return float(value)
    except (ValueError, TypeError):
        # Si hay un error en la conversión, asumir que es no numérico y convertir a 0
        return 0

# Aplicar la función a la columna 'price'
dfgames['price'] = dfgames['price'].apply(convert_to_zero_if_non_numeric)

# Realizar análisis estadístico en la columna 'price' actualizada
print(dfgames['price'].describe())
```
